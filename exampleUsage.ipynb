## Creating a Single Well
### Create the well first
Create a well and specify parameters if known

```python
G881 = Well(name='G-88-1', elevation=233.84, project='06-0899-266-01', logged_by='TW', date=1988,
           xcoord=628623.92, ycoord=5534173.82)
```

### Add intervals to the created well
Add intervals to the well specifying top and bottom and optionally adding a description

Note:
- Intervals *do not* have to be added in a top to bottom order
- Lithologies are currently limited to those in the plotting dictionary (see code)
- Descriptions in the middle of a unit may be added by simply breaking the interval into different sections (see 'Clay' unit below)

```python
G881.add_interval(top=0, bottom=0.75, lithology='Fill', description='Topsoil with clay and granular fill below')
G881.add_interval(top=0.75, bottom=1.75, lithology='Silt', description='Silt; Tan and damp')
G881.add_interval(top=1.75, bottom=5, lithology='Clay', description='Clay; Brown, stiff, with trace silt and minor inclusions')
G881.add_interval(top=5, bottom=8.5, lithology='Clay', description='Clay; Grey, highly plastic, with pockets of till increasing in size with depth')
G881.add_interval(top=8.5, bottom=10.75, lithology='Till', description='Till; Grey and damp with silt and clay, becoming granular with cobbles at depth')
```

### Print a summary of the well

```python
G881.summary()
```

### Remove intervals
If a mistake is made, intervals can be removed either by layer number (from surface) or by entire lithology

```python
G881.remove_interval(layNum=1)
G881.summary()
```

```python
G881.remove_interval(lithology='Clay')
G881.summary()
```

The last example above shows how removing using the `lithology` method can be dangerous, as it will remove *all* layers with that lithology

## Plotting a Single Well
A single well can be plotted quickly.  Note that colors and hatches can easily be changed by manipulating the plotting dictionary within the code.

```python
#add back removed intervals
G881.add_interval(top=0.75, bottom=1.75, lithology='Silt', description='Silt; Tan and damp')
G881.add_interval(top=1.75, bottom=5, lithology='Clay', description='Clay; Brown, stiff, with trace silt and minor inclusions')
G881.add_interval(top=5, bottom=8.5, lithology='Clay', description='Clay; Grey, highly plastic, with pockets of till increasing in size with depth')
```

```python
G881.plotWell()
```
